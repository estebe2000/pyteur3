# Synchronicité et asynchronicité

## `basthon.sleep`

```javascript
for(let i=1; i<=10; i++) {
    basthon.sleep(1);
    console.log(`${i*10}%`);
}
```

## Requête `fetch`

```javascript
// si la sortie est une promesse, elle est attendue

(async () => {
    // un peu d'attente
    await new Promise((resolve) => setTimeout(resolve, 1000))
    const response = await fetch("https://files.basthon.fr/hello.txt");
    if(response.ok)
        console.log(await response.text());
    else
        console.error("request failed!");
})();
```

## `prompt`

```javascript
function nombreMystere() {
    const mystere = Math.round(Math.random() * 100);
    let m;
    
    while ((m = parseInt(prompt("Propose un nombre entre 0 et 100 : "))) !== mystere) {
        if (m > mystere)
            console.log("trop grand");
        else
            console.log("trop petit");
    }
    console.log("Bravo ! C'est bien ça !");
}

nombreMystere();
```
