## Dessiner sur un canvas

```javascript
const w = 720, h = 480;

const canvas = new Canvas(w, h);
const ctx = canvas.getContext("2d");
const img = ctx.createImageData(w, h);

for (let i = 0; i < w; i++)
    for (let j = 0; j < h; j++) {
        const p = 4 * (j * w + i);
        img.data[p + 0] = (255 * (h - 1 - j)) / (h - 1); // red
        img.data[p + 1] = (255 * (w - 1 - i)) / (w - 1); // green
        img.data[p + 2] = (255 * j) / (h - 1);           // blue
        img.data[p + 3] = 255;                           // alpha
    }

ctx.putImageData(img, 0, 0);

basthon.display(canvas);
```

```javascript
const w = 720, h = 480;

const canvas = new Canvas(w, h);
const ctx = canvas.getContext("2d");

const colors = ['#046975', '#2EA1D4', '#3BCC2A', '#FFDF59', '#FF1D47'];

for (let i = 0; i < 1000; i++) {
    const radius = Math.random() * 20 + 1;
    const x = Math.random() * (w - radius  * 2) + radius;
    const y = Math.random() * (h - radius  * 2) + radius;
    
    ctx.beginPath();
    ctx.arc(x, y, radius, 0, Math.PI * 2, false);
    ctx.fillStyle = colors[i % colors.length];
    ctx.fill();
}

basthon.display(canvas);
```
