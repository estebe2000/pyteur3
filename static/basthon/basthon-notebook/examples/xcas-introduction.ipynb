## Un peu d'algèbre linéaire

```xcas
A := [[1, 0, 1],
      [1, 1, 1],
      [0, 1, 0]]
```

```xcas
ker(A)
```

```xcas
image(A)
```

```xcas
factor(charpoly(A, X))
```

```xcas
P := [[1, 1, 0],
      [0, 2, 2],
      [-1, 1, 0]]
```

```xcas
M := inv(P) * A * P
```

```xcas
U := [[x], [y], [z]]
```

```xcas
solve(A * U = 2U, [x, y, z])
```
